```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../src/config/database.db")

df_levels = pd.read_sql_query("""
    SELECT 
        level,
        COUNT(*) as cantidad,
        ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM course_recommendation), 2) as porcentaje
    FROM course_recommendation
    GROUP BY level
    ORDER BY level
""", conn)

display(df_levels)
```

```python
for level in ['Advanced', 'Intermediate', 'Introductory']:
    df_sample = pd.read_sql_query(f"""
        SELECT title, level, subject
        FROM course_recommendation
        WHERE level = '{level}'
        LIMIT 3
    """, conn)
    
    print(f"\n{level.upper()} ({len(df_sample)} ejemplo):")
    print(df_sample)
```
